```python
import os
import pandas as pd
from IPython.display import display_html 
```

```python
[f for f in os.listdir() if f.endswith(".json")]
```

```python
df_dict = {}
df_dict["LSTM"] = "cs230_classification_report_LSTM_local.json"
df_dict["scratch_softmax_contrastive"] = "cs230_classification_report_scratch_softmax_contrastive.json"
df_dict["scratch_softmax_only"] = "cs230_classification_report_scratch_softmax_only.json"
df_dict["tuning_softmax_only"] = "cs230_classification_report_tuning_softmax.json"
df_dict["tuning_softmax_contrastive"] = "cs230_classification_report_tuning_softmax_contrastive.json"

df_dict = {key:pd.read_json(value) for key, value in df_dict.items()}

cols = ["index", "f1-score", "precision", "recall"]

df_dict = {key:value[cols].set_index("index") for key, value in df_dict.items()}
df_dict = {key:value.loc[["accuracy", "macro avg", "weighted avg"]] for key, value in df_dict.items()}
```

```python
# from: https://blog.softhints.com/display-two-pandas-dataframes-side-by-side-jupyter-notebook/

df0_styler = df_dict["LSTM"].loc[["accuracy", "macro avg", "weighted avg"]].style.set_table_attributes("style='display:inline'").set_caption("Multilayer Bi-LSTM")

df1_styler = df_dict["tuning_softmax_only"].loc[["accuracy", "macro avg", "weighted avg"]].style.set_table_attributes("style='display:inline'").set_caption("Tuned: Softmax Only")
df2_styler = df_dict["tuning_softmax_contrastive"].loc[["accuracy", "macro avg", "weighted avg"]].style.set_table_attributes("style='display:inline'").set_caption("Tuned: Softmax + Contrastive Loss")

df3_styler = df_dict["scratch_softmax_only"].loc[["accuracy", "macro avg", "weighted avg"]].style.set_table_attributes("style='display:inline'").set_caption("New: Softmax Only")
df4_styler = df_dict["scratch_softmax_contrastive"].loc[["accuracy", "macro avg", "weighted avg"]].style.set_table_attributes("style='display:inline'").set_caption("New: Softmax + Contrastive Loss")

display_html(df0_styler._repr_html_(), raw=True)
print()
display_html(df1_styler._repr_html_()+ " " + df2_styler._repr_html_(), raw=True)
print()
display_html(df3_styler._repr_html_()+ " " + df4_styler._repr_html_(), raw=True)
```
